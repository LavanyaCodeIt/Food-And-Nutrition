```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv("Recipe_clean.csv")
#Histogram on Bivariate Variable
# Plotting a Histogram
data.Diet.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Diet vs Total Cooking Time")
plt.ylabel("Total Cooking Time/Min")
plt.xlabel("Diet")
```

```python
data.Course.value_counts().nlargest(10).plot(kind="bar", figsize=(10,5))
plt.title("Course Vs Total Time")
plt.ylabel("Total Time/Min")
plt.xlabel("Course Types")
```

```python
data.Cuisine.value_counts().nlargest(50).plot(kind="bar",figsize=(30,25))
plt.title("Time Taken for each Cuisine")
plt.xlabel("Cuisine")
plt.ylabel("Cooking Time/Min")
```

```python
#plot the scatter plot of Cooking time and Preparation Time variable in data (Bivariate Analysis)
x=data["Preparation Time/Min"]
y=data["Cooking Time/Min"]
plt.scatter(x,y)
plt.title("Preparation Time vs Cooking Time")
plt.xlabel("Preparation Time")
plt.ylabel("Cooking Time")
plt.show()
```

```python
#plot the scatter plot of Total Time  and  Servings variable in data(bivariate :numeric-numeric analysis)
plt.scatter(x=data["Servings"],y=data["Total Time/Min"])
plt.title("Total Time based on Servings")
plt.xlabel("Total Time")
plt.ylabel("Servings")
plt.show()
```

```python
#plot the pair plot of prep,cooking and servings data dataframe.
sns.pairplot(data = data, vars=['Preparation Time/Min','Cooking Time/Min','Servings'])
plt.show()
```

```python
#Plotting Categorical Ordered Variable on Course
data.Course.value_counts(normalize=True)

#plot the pie chart of Course categories
data.Course.value_counts(normalize=True).plot.barh()
plt.show()
```

```python
#Plotting Categorical UnOrdered Univariate on Course
data.Diet.value_counts(normalize=True)

#plot the pie chart of Course categories
data.Diet.value_counts(normalize=True).plot.barh()
```

```python
#creating a matrix for more than 2 variable ,all the variable here are numeric
corr_numeric=data[['Preparation Time/Min','Cooking Time/Min','Servings']].corr()

#plot the correlation matrix of Preparation Time/Min','Cooking Time/Min','Servings in data dataframe.
sns.heatmap(data[['Preparation Time/Min','Cooking Time/Min','Servings']].corr(), annot=True, cmap = 'Reds')
plt.show()
print("The correlation :", corr_numeric)
```

```python
#detecting outliers using boxplot- univariate
sns.boxplot(x=data['Preparation Time/Min'])
```

```python
#detecting outliers- Bivariate
sns.boxplot(x=data['Cooking Time/Min'],y=data['Servings'])
```
