```python
# import libraries
import pandas as pd
```

```python
# load data
df = pd.read_csv('Recipe.csv')
df.head()
```

```python
# keep only English columns (easy for string operations)
df.drop(['RecipeName', 'Ingredients', 'Instructions'], axis=1, inplace=True)

# rename some columns for easy reference
df = df.rename(columns={'TranslatedRecipeName': 'RecipeName',
                        'TranslatedIngredients': 'Ingredients', 
                        'TranslatedInstructions': 'Instructions'})

df.head()
```

```python
# sort based on time to find why some recipes had 0 time. 
df.sort_values('TotalTimeInMins').iloc[0].URL

'''Got the URL and looked at it manually. 
Looks like the data is not entered correctly for this recipe. 
We could manually read the page and correct our data'''
```

```python
# looking to see what categories are present in Cuisine
df.Cuisine.unique()
```

```python
# some categories in cuisines doesn't make sense such as snack, dinner, brunch, lunch, appetizer. 
# These are the courses so want to analyze those recipes

temp_list = ['dinner', 'lunch', 'brunch', 'snack', 'appetizer', 'dessert']
temp_pattern = '|'.join(temp_list)
df[df.Cuisine.str.lower().str.contains(temp_pattern)]

'''looks like the course information can be found both in Cuisine and Course columns. 
When cuisine has course info, the cuisine is unknown (I looked at some links and found 
things for example by running "df.loc[3118].URL" and found recipes like chips 
which are not really conneced o any cuisine. I also found other recipes
that says Moroccan fish recipe - it should be labeled Moroccan in Cuisine but its not done.
Cuisine may be found in RecipeName column'''
```

```python
# looking to see what categories are present in Course
df.Course.unique()
```

```python
# looking to see what categories are present in Diet
df.Diet.unique()
```

```python
# I see many categorical variables so want to visualize it for 2 variables at a time using stack bar graph 

df_stackplot = df.groupby(by=['Course', 'Diet']).size()
df_stackplot.unstack().plot.bar(stacked=True, figsize=(20,10))
```

```python
# In the above plot, I am curious why there are non vegetarian items under Veg Courses. 
# So I am gonna look at one recipe for which this happens
df[( (df.Course == 'Vegetarian') & (df.Diet.str.contains('Non')) )]

'''it looks like these are actually vegetarian recipes but 
are often recommended/included in some non vegetarian diets'''
```

```python
# Lets check the other way around - where we see a vegetarian diet and see if any of those has a non veg recipe. 
# So I am gonna look at one recipe for which this happens
df[( (df.Diet == 'Vegetarian') & (df.Course.str.contains('Non')) )]

'''it looks like these are non-vegetarian recipes but 
they are recommended/included in a vegetarian diet. This certainly doesn't make sense'''
```
